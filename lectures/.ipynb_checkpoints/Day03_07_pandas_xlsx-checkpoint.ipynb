```python
# import 
import numpy as np
import pandas as pd
```

## 데이터 적재 및 통합

```python
# xlsx format read_excel로 읽기
data = pd.read_excel('../data/NC Dinos.xlsx',sheet_name=None)
NC13, NC14, NC15 = data.values()
```

```python
NC13.head()
```

```python
NC13 = NC13.drop('Unnamed: 0',axis=1)
NC13 = NC14.drop('Unnamed: 0',axis=1)
NC15.drop('Unnamed: 0',axis=1,inplace = True)
```

```python
NC13['연도'] = 2013
NC14['연도'] = 2014
NC15['연도'] = 2015
```

```python
NCAll = pd.concat([NC13,NC14,NC15],axis=0,ignore_index=True)
```

```python
NCAll.head()
```

## 2015년도 선수 기록을 '타율'로 내림차순 정렬

```python
NC15.sort_values(by='타율',ascending=False)
```
